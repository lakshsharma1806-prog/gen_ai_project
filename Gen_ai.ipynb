```python
import sys
!{sys.executable} -m pip install streamlit
import streamlit as st
from generator import generate_content
from download import get_txt_download, get_pdf_download

# ── Page config ──────────────────────────────────────────────────────────────
st.set_page_config(
    page_title="AI Creative Writer",
    page_icon="✍️",
    layout="centered",
    initial_sidebar_state="collapsed",
)

# ── Custom CSS ────────────────────────────────────────────────────────────────
st.markdown("""
<style>
@import url('https://fonts.googleapis.com/css2?family=Playfair+Display:wght@600;700&family=Inter:wght@400;500&display=swap');

html, body, [class*="css"] {
    font-family: 'Inter', sans-serif;
}
h1, h2, h3 {
    font-family: 'Playfair Display', serif !important;
}
.main { background: #0f0f0f; }
.stApp { background: linear-gradient(135deg, #0f0f1a 0%, #1a1a2e 100%); }

.hero-title {
    font-family: 'Playfair Display', serif;
    font-size: 2.8rem;
    font-weight: 700;
    background: linear-gradient(135deg, #e8d5b7, #f5c842, #e8d5b7);
    -webkit-background-clip: text;
    -webkit-text-fill-color: transparent;
    text-align: center;
    line-height: 1.2;
    margin-bottom: 0.3rem;
}
.hero-sub {
    text-align: center;
    color: #888;
    font-size: 1rem;
    margin-bottom: 2rem;
}
.output-box {
    background: #1e1e2e;
    border: 1px solid #333;
    border-left: 4px solid #f5c842;
    border-radius: 8px;
    padding: 1.5rem 1.8rem;
    color: #e0e0e0;
    font-size: 1rem;
    line-height: 1.9;
    white-space: pre-wrap;
    font-family: 'Georgia', serif;
    margin-top: 1rem;
}
.word-count {
    color: #888;
    font-size: 0.8rem;
    text-align: right;
    margin-top: 0.4rem;
}
div[data-testid="stTabs"] button {
    font-family: 'Playfair Display', serif !important;
    font-size: 1rem !important;
    color: #aaa !important;
}
div[data-testid="stTabs"] button[aria-selected="true"] {
    color: #f5c842 !important;
    border-bottom: 2px solid #f5c842 !important;
}
.stButton > button {
    background: linear-gradient(135deg, #f5c842, #e8a020) !important;
    color: #0f0f1a !important;
    font-weight: 600 !important;
    border: none !important;
    border-radius: 6px !important;
    padding: 0.6rem 2rem !important;
    font-size: 1rem !important;
    transition: opacity 0.2s !important;
}
.stButton > button:hover { opacity: 0.88 !important; }
.stDownloadButton > button {
    background: transparent !important;
    color: #f5c842 !important;
    border: 1px solid #f5c842 !important;
    border-radius: 6px !important;
    font-size: 0.85rem !important;
}
label, .stSelectbox label, .stSlider label, .stTextInput label, .stTextArea label {
    color: #ccc !important;
    font-size: 0.9rem !important;
}
.stTextInput input, .stTextArea textarea {
    background: #1e1e2e !important;
    color: #e0e0e0 !important;
    border: 1px solid #333 !important;
    border-radius: 6px !important;
}
.stSelectbox > div > div {
    background: #1e1e2e !important;
    color: #e0e0e0 !important;
    border: 1px solid #333 !important;
}
.stSlider > div > div > div { background: #f5c842 !important; }
</style>
""", unsafe_allow_html=True)

# ── Header ────────────────────────────────────────────────────────────────────
st.markdown('<div class="hero-title">✍️ AI Creative Writer</div>', unsafe_allow_html=True)
st.markdown('<div class="hero-sub">Generate stories, essays & poems powered by AI</div>', unsafe_allow_html=True)

# ── API Key input (sidebar) ───────────────────────────────────────────────────
with st.sidebar:
    st.markdown("### ⚙️ Settings")
    api_key = st.text_input("OpenAI API Key", type="password", placeholder="sk-...")
    st.markdown("---")
    st.markdown("**Model**")
    model = st.selectbox("", ["gpt-3.5-turbo", "gpt-4", "gpt-4o"], label_visibility="collapsed")
    st.markdown("---")
    temperature = st.slider("", 0.1, 1.5, 0.85, 0.05, label_visibility="collapsed")
    st.caption("Higher = more creative / unpredictable")
    st.markdown("---")
    st.caption("Built for BTech Generative AI Project")

# ── Tabs ──────────────────────────────────────────────────────────────────────
tab1, tab2, tab3 = st.tabs(["📖  Story", "📝  Essay", "🌸  Poem"])

def show_result(result, filename_base):
    wc = len(result.split())
    st.markdown(f'<div class="output-box">{result}</div>', unsafe_allow_html=True)
    st.markdown(f'<div class="word-count">~{wc} words</div>', unsafe_allow_html=True)
    col1, col2 = st.columns(2)
    with col1:
        st.download_button("⬇️ Download .txt", get_txt_download(result),
                           file_name=f"{filename_base}.txt", mime="text/plain")
    with col2:
        st.download_button("📄 Download .pdf", get_pdf_download(result, filename_base),
                           file_name=f"{filename_base}.pdf", mime="application/pdf")

# ── STORY TAB ─────────────────────────────────────────────────────────────────
with tab1:
    st.markdown("#### Craft your story")
    c1, c2 = st.columns(2)
    with c1:
        story_topic = st.text_input("Topic / Title", placeholder="e.g. A lost astronaut finds home", key="s_topic")
    with c2:
        story_genre = st.selectbox("Genre", ["Fantasy", "Mystery", "Sci-Fi", "Romance",
                                             "Horror", "Adventure", "Historical Fiction", "Thriller"], key="s_genre")
    c3, c4 = st.columns(2)
    with c3:
        story_tone = st.selectbox("Tone", ["Neutral", "Dark & Gritty", "Light & Humorous",
                                           "Emotional & Dramatic", "Suspenseful", "Whimsical"], key="s_tone")
    with c4:
        story_length = st.slider("Word count", 150, 1000, 350, 50, key="s_len")

    story_extra = st.text_area("Additional instructions (optional)",
                               placeholder="e.g. Include a plot twist. The hero is a 12-year-old girl.",
                               height=80, key="s_extra")
    if st.button("✨ Generate Story", key="s_gen"):
        if not api_key:
            st.error("Please enter your OpenAI API key in the sidebar.")
        elif not story_topic.strip():
            st.warning("Please enter a topic for your story.")
        else:
            with st.spinner("Crafting your story..."):
                result = generate_content(
                    api_key, model, temperature,
                    content_type="story",
                    topic=story_topic,
                    style=story_genre,
                    tone=story_tone,
                    length=story_length,
                    extra=story_extra,
                )
            if result.startswith("ERROR:"):
                st.error(result)
            else:
                st.session_state["story_result"] = result

    if "story_result" in st.session_state:
        show_result(st.session_state["story_result"], "my_story")
        if st.button("🔄 Continue story", key="s_cont"):
            with st.spinner("Continuing..."):
                continuation = generate_content(
                    api_key, model, temperature,
                    content_type="story_continuation",
                    topic=story_topic,
                    style=story_genre,
                    tone=story_tone,
                    length=250,
                    extra=st.session_state["story_result"][-500:],
                )
            if not continuation.startswith("ERROR:"):
                st.session_state["story_result"] += "\n\n" + continuation
                st.rerun()

# ── ESSAY TAB ─────────────────────────────────────────────────────────────────
with tab2:
    st.markdown("#### Compose your essay")
    c1, c2 = st.columns(2)
    with c1:
        essay_topic = st.text_input("Essay topic", placeholder="e.g. The impact of social media on youth", key="e_topic")
    with c2:
        essay_type = st.selectbox("Essay type", ["Argumentative", "Descriptive",
                                                  "Expository", "Narrative", "Persuasive", "Analytical"], key="e_type")
    c3, c4 = st.columns(2)
    with c3:
        essay_level = st.selectbox("Reading level", ["High School", "Undergraduate", "Postgraduate", "Professional"], key="e_lvl")
    with c4:
        essay_length = st.slider("Word count", 200, 1200, 500, 50, key="e_len")

    essay_stance = st.text_input("Your stance / thesis (optional)",
                                 placeholder="e.g. Social media has a net negative effect on teenage mental health", key="e_stance")
    if st.button("✨ Generate Essay", key="e_gen"):
        if not api_key:
            st.error("Please enter your OpenAI API key in the sidebar.")
        elif not essay_topic.strip():
            st.warning("Please enter an essay topic.")
        else:
            with st.spinner("Writing your essay..."):
                result = generate_content(
                    api_key, model, temperature,
                    content_type="essay",
                    topic=essay_topic,
                    style=essay_type,
                    tone=essay_level,
                    length=essay_length,
                    extra=essay_stance,
                )
            if result.startswith("ERROR:"):
                st.error(result)
            else:
                st.session_state["essay_result"] = result

    if "essay_result" in st.session_state:
        show_result(st.session_state["essay_result"], "my_essay")

# ── POEM TAB ──────────────────────────────────────────────────────────────────
with tab3:
    st.markdown("#### Compose your poem")
    c1, c2 = st.columns(2)
    with c1:
        poem_theme = st.text_input("Theme / subject", placeholder="e.g. The first rain of monsoon", key="p_theme")
    with c2:
        poem_style = st.selectbox("Style", ["Free Verse", "Haiku", "Sonnet",
                                            "Limerick", "Ballad", "Ode", "Acrostic", "Villanelle"], key="p_style")
    c3, c4 = st.columns(2)
    with c3:
        poem_mood = st.selectbox("Mood", ["Joyful", "Melancholic", "Romantic", "Reflective",
                                          "Hopeful", "Dark", "Playful", "Spiritual"], key="p_mood")
    with c4:
        poem_variants = st.slider("Number of variations", 1, 3, 1, key="p_var")

    if st.button("✨ Generate Poem", key="p_gen"):
        if not api_key:
            st.error("Please enter your OpenAI API key in the sidebar.")
        elif not poem_theme.strip():
            st.warning("Please enter a theme for your poem.")
        else:
            poems = []
            with st.spinner(f"Composing {poem_variants} poem(s)..."):
                for i in range(poem_variants):
                    result = generate_content(
                        api_key, model, min(temperature + i * 0.1, 1.5),
                        content_type="poem",
                        topic=poem_theme,
                        style=poem_style,
                        tone=poem_mood,
                        length=120,
                        extra="",
                    )
                    poems.append(result)
            st.session_state["poem_results"] = poems

    if "poem_results" in st.session_state:
        poems = st.session_state["poem_results"]
        if len(poems) == 1:
            show_result(poems[0], "my_poem")
        else:
            for i, poem in enumerate(poems, 1):
                st.markdown(f"**Variation {i}**")
                show_result(poem, f"my_poem_v{i}")
                st.markdown("---")
```

```python
%%writefile generator.py
import openai

def generate_content(api_key, model, temperature, content_type, topic, style, tone, length, extra):
    try:
        client = openai.OpenAI(api_key=api_key)

        # Construct prompt based on content_type
        prompt = f"Write a {content_type} about '{topic}' in a {style} style with a {tone} tone, approximately {length} words long. "
        if extra:
            prompt += f"Additional instructions: {extra}"

        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "system", "content": "You are a creative writing assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=temperature,
            max_tokens=length * 2 # Roughly double for token to word conversion
        )
        return response.choices[0].message.content
    except openai.APIError as e:
        return f"ERROR: OpenAI API Error: {e}"
    except Exception as e:
        return f"ERROR: An unexpected error occurred: {e}"


def generate_content_continuation(api_key, model, temperature, content_type, topic, style, tone, length, existing_content):
    try:
        client = openai.OpenAI(api_key=api_key)

        prompt = f"Continue the following {content_type} about '{topic}' in a {style} style with a {tone} tone, adding approximately {length} words. The existing content is: \n\n{existing_content}\n\nContinue from here:"

        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "system", "content": "You are a creative writing assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=temperature,
            max_tokens=length * 2
        )
        return response.choices[0].message.content
    except openai.APIError as e:
        return f"ERROR: OpenAI API Error: {e}"
    except Exception as e:
        return f"ERROR: An unexpected error occurred: {e}"
```

```python
%%writefile download.py
from fpdf import FPDF

def get_txt_download(text_content):
    return text_content.encode('utf-8')

def get_pdf_download(text_content, filename_base):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)
    pdf.multi_cell(0, 10, text_content)
    return pdf.output(dest='S').encode('latin-1')
```
